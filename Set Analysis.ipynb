```python
%matplotlib inline
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
def get_classes(path):
    classes = {arch_class: len(os.listdir(os.path.join(path, arch_class)))  for arch_class in os.listdir(path) if os.path.isdir(os.path.join(path, arch_class))}
    print("Classes: {num_classes}".format(num_classes = len(classes)))
    print("Images: {num_img}".format(num_img = sum(classes.values())))
    print()
    for key in classes:
        print("{class_name}: {class_count}".format(class_name=key, class_count=classes[key]))
    return classes
```

```python
set_folder = "c:\\Users\\Tomasz Marcinkowski\\Documents\\MEGA\\Zbiory\\"
set_1_path = os.path.join(set_folder, "Zbior3_koledzy\\")
set_2_path = os.path.join(set_folder, "Zbior3_koledzy_transformed\\")
print("Set 1:")
classes_1 = get_classes(set_1_path)
print()
print("Set 2:")
classes_2 = get_classes(set_2_path)
```

```python
plt.rcParams['figure.figsize'] = (15,9)
N = len(classes_1)

ind = np.arange(N)
width = 0.35

fig, ax = plt.subplots()
rects1 = ax.bar(ind, classes_1.values(), width, color='r')
rects2 = ax.bar(ind + width, classes_2.values(), width, color='y')

ax.set_ylabel('Number of images')
ax.set_xticks(ind + width / 2)
ax.set_xticklabels(classes_1.keys(), rotation = 90)


ax.legend((rects1[0], rects2[0]), ('Set1', 'Set2'))


plt.show()
```
