```python
%matplotlib inline
import os
import matplotlib.pyplot as plt
```

```python
set_folder = "c:\\Users\\Tomasz Marcinkowski\\OneDrive\\Magisterskie\\Semestr III\\Nowe trendy w obliczeniach neuronowych\\Projekt\\Zbiory\\"
set_1_path = os.path.join(set_folder, "Zbiór1\\")
classes = {arch_class: len(os.listdir(os.path.join(set_1_path, arch_class)))  for arch_class in os.listdir(set_1_path) if os.path.isdir(os.path.join(set_1_path, arch_class))}
print("Classes: {num_classes}".format(num_classes = len(classes)))
print("Images: {num_img}".format(num_img = sum(classes.values())))
print()
for key in classes:
    print("{class_name}: {class_count}".format(class_name=key, class_count=classes[key]))
```

```python
plt.bar(range(len(classes)), list(classes.values()), align='center')
plt.xticks(range(len(classes)), list(classes.keys()), rotation = 90)
plt.ylabel('Wartość')
plt.rcParams['figure.figsize'] = (10, 6)
plt.show()
```
